```python
columns_to_remove = ['Metadata', 'Time']
rdata.drop(columns=columns_to_remove, inplace=True)
rdata.columns
```

```python
import re
import string
```

```python
def textclean1(text):
    if isinstance(text,str):
        text=text.lower()
        text=re.sub('\[.*?\]','',text)
        text=re.sub('[%s]'%re.escape(string.punctuation),'',text)
        text=re.sub('\w*\d\w*','',text)
        return text
    else:
        return text
cleaned1=lambda x: textclean1(x)
```

```python
#updated text
rdata['cleaned description']=pd.DataFrame(rdata.Review.apply(cleaned1))
rdata.columns
```

```python
#removing new line char
def textclean2(text):
    if isinstance(text, str):
        text = re.sub(r'[\'"…]', '', text)
        text = re.sub(r'\n', '', text)  # Replace newline characters
        return text
    else:
        return text

cleaned2=lambda x: textclean2(x)
```

```python
rdata['cleaned_description_new']=pd.DataFrame(rdata['cleaned description'].apply(cleaned2))
rdata.head()
```

```python
from textblob import TextBlob
```

```python
def analyze_sentiment(cleaned_description_new):
    if isinstance(cleaned_description_new, str):
        analysis = TextBlob(cleaned_description_new)
        # Classify as positive if sentiment polarity is >= 0, else classify as negative
        return "Positive" if analysis.sentiment.polarity >= 0 else "Negative"
    else:
        return "Neutral"
```

```python
rdata['predicted_label'] = rdata['cleaned_description_new'].apply(analyze_sentiment)
```
